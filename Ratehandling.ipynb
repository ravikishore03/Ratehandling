```python
import time

requests = {}
LIMIT = 5
TIME_WINDOW = 60


def is_allowed(user_id):
    current_time = time.time()
    if user_id not in requests:
        requests[user_id] = []

    # Remove timestamps that are outside the time window
    requests[user_id] = [timestamp for timestamp in requests[user_id] if current_time - timestamp < TIME_WINDOW]

    if len(requests[user_id]) < LIMIT:
        requests[user_id].append(current_time)
        return True
    else:
        return False

    # Testing
    for i in range(7):
        if is_allowed("user1"):
          print(i+1,is_allowed("user123"))
```
